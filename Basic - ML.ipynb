```python
iris = datasets.load_iris()
train = iris.data
label = iris.target
```

```python
train_feature_big,test_feature,train_label_big,test_label = train_test_split(train,label,test_size=0.2,random_state=1)
train_feature,dev_feature,train_label,dev_label = train_test_split(train_feature_big,train_label_big,test_size=0.25,random_state=1)
```

```python
svc = SVC(C=100.0,gamma=1,tol=0.0001,kernel='rbf',degree=3, coef0=0.0,shrinking=True,probability=False,class_weight ="balanced",verbose=False,cache_size=200,max_iter=-1, decision_function_shape='ovr', random_state=Seed)
knn = KNeighborsClassifier(n_neighbors=5, weights='uniform', algorithm='auto',leaf_size=30, p=2, metric='minkowski', metric_params=None)
LR = LogisticRegression(penalty="l2", dual=False, tol=0.0001, C=1.0, fit_intercept=True, intercept_scaling=1, class_weight="balanced", random_state=Seed, max_iter=100,  verbose=0, warm_start=False, n_jobs=None)
DT= DecisionTreeClassifier(random_state=Seed) 
```

```python
svc.fit(train_feature,train_label)
pred = svc.predict(dev_feature)
print(classification_report(dev_label, pred))
```

```python
svc.set_params(C=150)
print(svc)
```

```python
list_C = [100,150]
list_gamma = [0.01,0.1]
list_tol = [0.001]

for C in list_C :
    for gamma in list_gamma:
        for tol in list_tol:
            svc.set_params(C=C)
            svc.set_params(gamma=gamma)
            svc.set_params(tol=tol)
            svc.fit(train_feature,train_label)
            pred = svc.predict(dev_feature)
            print(svc)
            print(classification_report(dev_label, pred))
```
