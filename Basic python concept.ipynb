# NO2 : String (字串) 一串字元

##### 去除字符

```python
string_hello = '   Hello world !,   '
print(string_hello.strip()) #strip()方法用于移除字符串头尾指定的字符（默认为空格）
print(string_hello.strip().lstrip()) # lstrip()方法用于截掉字符串左边的空格或指定字符
print(string_hello.strip().lstrip().rstrip(',')) # rstrip()方法用于截掉字符串右边的空格或指定字符
```

##### 插入字符

```python
string_hello_clear = string_hello.strip().lstrip().rstrip(',')
string_hello_clear_add = string_hello_clear.join(['start ~ ',' and ',' ~ end'])
print(string_hello_clear_add)
```

##### 搜尋字符

```python
hello_position = string_hello_clear_add.find('Hello')

print('position : ',hello_position)
print(string_hello_clear_add[hello_position:hello_position + len('Hello')])
```

##### 比較字符

```python
print(string_hello >  string_hello_clear_add)
print(string_hello ==  string_hello_clear_add)
print(string_hello <  string_hello_clear_add)
```

```python
len(string_hello_clear_add)
```

##### 大小寫改變

```python
string_hello_clear_add.upper()
```

```python
string_hello_clear_add.lower()
```

```python
string_hello_clear_add.swapcase()
```

```python
string_hello_clear_add.capitalize()
```

##### 切割字串

```python
string_hello_clear_add.split('and')
```

```python
len(list_A) #長度
```

```python
max(list_A) , min(list_A) # 最大最小值
```

##### 從最後面加入值 、 計算出現次數 、 加入另一個列表

```python
list_A.append(100)
print(list_A)
```

```python
list_A.append(9)
print(list_A.count(9))
```

```python
list_A.extend(list_A)
print(list_A)
```

```python
list_A.insert(2,100)
print(list_A)
```

##### 搜尋第一個符合的元素 、 移除列表的元素

```python
print(list_A.index(100))
```

```python
list_A.pop() #移除列表的一個元素 預設是最後一個元素
```

```python
list_A.pop(0) #移除位置
```

```python
list_A
```

```python
list_A.remove(100) #移除列表符合的第一個元素
print(list_A)
```

##### 反轉列表 、 排序列表

```python
list_A.reverse()
print(list_A)
```

```python
list_A.sort()
print(list_A)
```

# NO4 : Tuple (元組)

#### 元組用()括住 內部元素以逗號隔開 
#### 是列表資料格式的簡化版本
#### 不能二次設定值 類似唯讀列表

```python
tuple_A = ('hello','world','!')
```

```python
print(tuple_A)
print(tuple_A[1:])
print(tuple_A[:1]+tuple_A[1:]+tuple_A[2:3]*4)
```

# NO5 : Dictionary (字典)

#### 字典用{}括住 由索引(key) & 對應值(value)組成

```python
dict_A = {}
dict_A['key_1'] = 'hello'
dict_A['key_2'] = 'world'
dict_A['key_3'] = '!'
```

```python
dict_A.keys()
```

```python
dict_B = {'key_1':'hello world !'}
```

```python
dict_A['key_1']
```

```python
dict_B
```

```python
dict_A.clear()
print(dict_A)
```

```python
dict_A = dict_B.copy()
print(dict_A)
```

```python
dict_A.get('key_1')
```

```python
if (dict_A.get('key_2') == None) : # 返回default
    print('None')
```

```python
dict_A.items()
```

```python
dict_A.setdefault('key_2')
print(dict_A)
```

```python
dict_B.update(dict_A)
print(dict_B)
```
