# Part2 :操作 cell 

## ESC / Ctrl-M 切回綠色 可做cell操作

* Up/k Down/j + shift(連續選擇)
* x : 剪下現在cell
* 連按D+D : 刪除cell 
* a : 在上方新增一個cell
* b : 在下方新增一個cell
* c : 複製選中的cell
* v : 往下黏貼
* z : 回復上一個刪除的cell
* shift+M : 合併選中的cell
* shift+space : 向上滾動
* space : 向下滾動 

## status (ESC / Ctrl-M)

* ctrl+m
* y : code
* m : markdown
* r : raw

## Code : 

* 執行並到下個cell shift+Enter 
* 執行此cell ctrl+Enter
* 執行並插入下個cell alt+Enter



# Part3 : 編輯模式

## Enter 轉入編輯模式

* tab : 往後縮排(code模式)
* shift-Tab :　往前縮排(code模式)
* home : 此行開頭
* ctrl+home : cell開頭
* end : 此行結尾
* ctrl+end : cell結尾
* ctrl&shift&- : 從游標處分割cell

# Part4 : Markdown 

## 大標題 文字前 + #字符
## 排序 + *
* #字符越多 標題字體大小越小
* #字符後記得留空格
* 方程式用 錢字符 包起來 
* ex : $f(x) = 2x^2 + 4x + 1$
* 超連結用 中括號&小括號
* 圖片用 !中括號&小括號

### 切換code to markdown 快捷鍵

ctrl+m + m
